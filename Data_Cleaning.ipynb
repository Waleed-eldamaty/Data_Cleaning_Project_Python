## Creating a Dummy Variable for the "Reason for Absence" Column

```python
reasons_columns=pd.get_dummies(df['Reason for Absence'])
reasons_columns.head()
```

```python
# To check whether there is one reason only per each row, we create the "Checking" column by summing along the row
# 0 means there are missing values, 1 means there is only one reason, 2 means there are multiple reasons
reasons_columns['Checking']=reasons_columns.sum(axis=1)
reasons_columns.head()
```

```python
# To check the sum of the "Checking" column
reasons_columns['Checking'].sum(axis=0)
```

```python
# To make sure that "1" is the only value in the "checking" column
reasons_columns['Checking'].unique()
```

```python
# After making sure there is only 1 reason per each row, we can drop the "Checking" column
reasons_columns=reasons_columns.drop(['Checking'],axis=1)
reasons_columns.head()
```

```python
# Droping Reason "0"
reasons_columns=pd.get_dummies(df['Reason for Absence'],drop_first=True)
reasons_columns.head()
```

## Grouping "Reason for Absence"

```python
df.columns.values
```

```python
reasons_columns.columns.values
```

```python
# Drop the "Reason for Absence" column since the dummy variable is already created
df = df.drop(['Reason for Absence'], axis = 1)
```

```python
# Grouping of Absence Reasons
# Reason_type_1= 1-14
# Reason_type_2= 15-17
# Reason_type_3= 18-21
# Reason_type_4= 22-28
```

```python
reasons_columns.loc[:, 1:14].head(20)
```

```python
# "0" indicates the reason for absence is not from type_1 group of reasons
# "1" indicates the reason for absence is from type_1 group of reasons
reasons_columns.loc[:, 1:14].max(axis=1).head(20) #summing along the rows
```

```python
# splitting them into 4 Data Frames
reason_type_1=reasons_columns.loc[:, 1:14].max(axis=1)
reason_type_2=reasons_columns.loc[:, 15:17].max(axis=1)
reason_type_3=reasons_columns.loc[:, 18:21].max(axis=1)
reason_type_4=reasons_columns.loc[:, 22:28].max(axis=1)
```

```python
reason_type_3.head(20)
```

## Concatenate Column Values

```python
# Adding the data frames together
df=pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)
df.head()
```

```python
# Changing the columns headers
df.columns.values
```

```python
Column_Names=['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns=Column_Names
df.head()
```

## Reordering Columns

```python
Column_Names_Reorderd=['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df=df[Column_Names_Reorderd]
df.head()
```

```python
df_copy=df.copy()
df_copy.head()
```

## Assigning "Timestamp" Value Type to the "Date" Column 

```python
df_copy['Date']=pd.to_datetime(df_copy['Date'], format = '%d/%m/%Y')
df_copy['Date'].head(20)
```

```python
df_copy.info()
```

## Extracting the "Month" value from the "Date" Column

```python
df_copy['Date'][0].month
```

```python
# Create an empty to list to store the values of "month" extracted from the column
list_months=[]
```

```python
# Extract the "Month" value from each row
for i in range(df_copy.shape[0]):
    list_months.append(df_copy['Date'][i].month)
```

```python
# Adding the "Month Value" as a new column
df_copy['Month Value']= list_months
df_copy.head(20)
```

## Extracting the "Day of the Week" from the "Date" Column

```python
df_copy['Date'][0].weekday()
```

```python
## Creating a function to be applied to each value in the "Date" column iteratively
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
# The "apply" function helps you apply the function you created repeatedly to the desired column (Date)
df_copy['Day of the Week']= df_copy['Date'].apply(date_to_weekday)
df_copy.head()
```

```python
# Dropping the "Date" column
df_copy=df_copy.drop(['Date'],axis=1)
df_copy
```

```python
df_copy.columns.values
```

```python
# Reordering the columns
Column_Names_Updated= ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4',  'Month Value',
       'Day of the Week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df_copy=df_copy[Column_Names_Updated]
df_copy.head()
```

## Converting "Education" Column into a Dummy Variable

```python
df_copy['Education'].unique()
```

```python
df_copy['Education'].value_counts()
```

```python
# 1 = highscool, 2= graduate, 3=postgraduate, 4= master or doctor
# Combining graduate and above into one category (1) while highschool education will be (0)
df_copy['Education']=df_copy['Education'].map({1:0, 2:1, 3:1, 4:1})
df_copy['Education'].value_counts()
```

```python
clean_df=df_copy.copy()
clean_df.head()
```
